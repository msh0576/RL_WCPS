```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.ticker as ticker
import pandas as pd
```

```python
def sliding_mean(data_array, window=20):
    data_array = np.array(data_array)
    new_list = []
    for i in range(len(data_array)):
        indices = range(max(i - window + 1, 0),
                        min(i + window + 1, len(data_array)))
        avg = 0
        for j in indices:
            avg += data_array[j]
        avg /= float(len(indices))
        new_list.append(avg)
        
    return np.array(new_list)

from scipy.interpolate import spline
r1 = pd.read_csv('./Origin.csv')
x1 = (r1["Step"] - r1["Step"][0])
r2 = pd.read_csv('./ACE.csv')
x2 = (r2["Step"] - r2["Step"][0])
```

```python
y1 = r1["Value"]
plt.plot(sliding_mean(x1), sliding_mean(y1), label='Origin', color = 'cornflowerblue')

y2 = r2["Value"]

plt.plot(sliding_mean(x2), sliding_mean(y2), label='ACE', color = 'tomato')

plt.legend() 

plt.plot(x1, y1, color = 'cornflowerblue', alpha=0.15)
plt.plot(x2, y2, color = 'tomato', alpha=0.25)
plt.xlabel('time(hours)')  
plt.ylabel('reward(meters advanced)') 
plt.grid(True) 
plt.savefig('relu-selu')
plt.title('Reward per test episode', size=14)
plt.show()
```
